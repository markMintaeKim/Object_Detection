```python
train_label_list = glob('yolov5/train_person/train/labels/*.txt')
print(len(train_label_list))
```

```python
rmv = 0
for path in train_label_list:
    f = open(path, "rt")
    cellphone = []
    for line in f:
        if line.startswith("21"):
            cellphone.append(line)
    f.close()
    if len(cellphone) == 0:
        os.remove(path)
        os.remove(path.replace(".txt", ".jpg").replace("/labels/", "/images/"))
        rmv += 1
    else:
        f2 = open(path, 'wt')
        for c in cellphone:
            f2.write(c.replace("21","0",1))
        f2.close()

print("Number of images left: {}".format(len(train_label_list)-rmv))
```

```python
# Relabing cellphone and person images into 0 and 1 respectively

rmv = 0
for path in train_label_list:
    f = open(path, "rt")
    person = []
    cellphone = []
    for line in f:
        if line.startswith("48"):
            person.append(line)
        if line.startswith("21"):
            cellphone.append(line)
    if len(person) == 0:
        os.remove(path)
        os.remove(path.replace(".txt", ".jpg").replace("/labels/", "/images/"))
        rmv += 1
    else:
        f2 = open(path, 'wt')
        for p in person:
            f2.write(p.replace("48","1",1))
        for c in cellphone:
            f2.write(c.replace("21","0",1))
        f2.close()
    f.close()
print("Number of images left: {}".format(len(train_label_list)-rmv))
```

```python
# Remove labels that are not in images and vice versa

from glob import glob
import os

train_image_list = glob('yolov5/dataset_3/mirror_selfie/images/*')
train_label_list = glob('yolov5/dataset_3/mirror_selfie/labels/*.txt')
print(len(train_image_list))
print(len(train_label_list))

for image in train_image_list:
    if not image.replace(".jpg", ".txt").replace("/images/", "/labels/") in train_label_list:
        train_image_list.remove(image)
        os.remove(image)

for label in train_label_list:
    if not label.replace(".txt", ".jpg").replace("/labels/", "/images/") in train_image_list:
        train_label_list.remove(label)
        os.remove(label)

print("Number of images left: {}".format(len(train_image_list)))
print("Number of labels left: {}".format(len(train_label_list)))
```

```python
# Train - Validation Split

import os
from glob import glob
import shutil
from sklearn.model_selection import train_test_split

image_files = glob('yolov5/dataset_3/train/images/*jpg')
images = [name.replace(".jpg", "") for name in image_files]

train_names, val_names = train_test_split(images, test_size=0.2, random_state=42, shuffle=True)

def batch_move_files(file_list, source_path, destination_path):
    for file in file_list:
        image = file.split('/')[-1] + '.jpg'
        txt = file.split('/')[-1] + '.txt'
        shutil.copy(os.path.join(source_path + '/images/', image), destination_path + '/images')
        shutil.copy(os.path.join(source_path + '/labels/', txt), destination_path + '/labels')
    return

source_dir = 'yolov5/train'
train_dir = 'yolov5/datasets_2/train'
val_dir = 'yolov5/datasets_2/val'

batch_move_files(train_names, source_dir, train_dir)
batch_move_files(val_names, source_dir, val_dir)
```
